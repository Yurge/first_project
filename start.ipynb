```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def matt(x):
    if x % 2 == 0:
        return -(x * 2)
    else:
        return x * 2
```

```python
X = [i for i in range(30)]
Y = [matt(i) for i in X]

plt.figure(figsize=(6, 2))
plt.plot(X, Y)
# plt.show()
```
